```
import numpy as np
import pandas as pd
df = pd.read_csv("Final_Data_v1.csv") 
len(df.userID.unique())
```

```
#Adding number of words feature
def no_of_words(summary):
  return len(summary.split()) 

df['TotalWords'] = df.apply (lambda row: no_of_words(row.summary), axis=1)
```

```
df.head()
```

```
#Adding complex words feature
def gt_6_words(s):
    words = s.split()
    return len([word for word in words if len(word)>6])

df['ComplexWords'] = df.apply (lambda row: gt_6_words(row.summary), axis=1)
df.head()
```

```
#Adding Unique words feature
def unique_words(summary):
  s = set()
  words = summary.split()
  for word in words:
    if len(word)>=5:
      s.add(word)
  return len(s)

df['UniqueWords'] = df.apply (lambda row: unique_words(row.summary), axis=1)
df.head()
```

```
#Adding average words per sentence 

import re
def avg_words_per_sentence(s):
  parts = [len(l.split()) for l in re.split(r'[?!.]', s) if l.strip()]
  return round(sum(parts)/len(parts),1)

df['AWPS'] = df.apply (lambda row: avg_words_per_sentence(row.summary), axis=1)
df.head()
```

```
def CountPOSFrequency(my_list):
  pos_freq = {}
  for items in my_list:
    pos_freq[items] = my_list.count(items)
  return pos_freq


def pos_tagging(summary):
  tokens = nltk.word_tokenize(summary)
  pos = nltk.pos_tag(tokens)
  pos_freq = []
  for i in range(len(pos)):
    pos_freq.append(pos[i][1])
  return CountPOSFrequency(pos_freq)

l = []
for index, row in df_summary.iterrows():
    l.append(pos_tagging(row['summary']))

print(l[0:10])
```

```
from nltk.corpus import cmudict
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from matplotlib import style
nltk.download('cmudict')
nltk.download('stopwords')
style.use("ggplot")
cmuDictionary = None

def syllable_count_Manual(word):
    word = word.lower()
    count = 0
    vowels = "aeiouy"
    if word[0] in vowels:
        count += 1
    for index in range(1, len(word)):
        if word[index] in vowels and word[index - 1] not in vowels:
            count += 1
            if word.endswith("e"):
                count -= 1
    if count == 0:
        count += 1
    return count

# Counts number of syllables:
def syllable_count(word):
    global cmuDictionary
    d = cmuDictionary
    try:
        syl = [len(list(y for y in x if y[-1].isdigit())) for x in d[word.lower()]][0]
    except:
        syl = syllable_count_Manual(word)
    return syl
  
# GIVES NUMBER OF SYLLABLES PER WORD
def Avg_Syllable_per_Word(text):
    tokens = word_tokenize(text, language='english')
    st = [",", ".", "'", "!", '"', "#", "$", "%", "&", "(", ")", "*", "+", "-", ".", "/", ":", ";", "<", "=", '>', "?",
          "@", "[", "\\", "]", "^", "_", '`', "{", "|", "}", '~', '\t', '\n']
    stop = stopwords.words('english') + st
    words = [word for word in tokens if word not in stop]
    syllabls = [syllable_count(word) for word in words]
    p = (" ".join(words))
    return sum(syllabls) / max(1, len(words))

dfinal['Avg_Syllables'] = dfinal.apply (lambda row: Avg_Syllable_per_Word(row.summary), axis=1)
dfinal.head()
```

```
def CountPunctuation(text):
    st = [",", ".", "'", "!", '"', ";", "?", ":", ";"]
    count = 0
    for i in text:
        if (i in st):
            count = count + 1
    return float(count) / float(len(text))

dfinal['Count_Punctuation'] = dfinal.apply (lambda row: CountPunctuation(row.summary), axis=1)
dfinal.head()
```

```
# RETURNS NORMALIZED COUNT OF FUNCTIONAL WORDS FROM A Framework for
# Authorship Identification of Online Messages: Writing-Style Features and Classification Techniques
def RemoveSpecialCHs(text):
    text = word_tokenize(text)
    st = [",", ".", "'", "!", '"', "#", "$", "%", "&", "(", ")", "*", "+", "-", ".", "/", ":", ";", "<", "=", '>', "?",
          "@", "[", "\\", "]", "^", "_", '`', "{", "|", "}", '~', '\t', '\n']

    words = [word for word in text if word not in st]
    return words
    
def CountFunctionalWords(text):
    functional_words = """a between in nor some upon
    about both including nothing somebody us
    above but inside of someone used
    after by into off something via
    all can is on such we
    although cos it once than what
    am do its one that whatever
    among down latter onto the when
    an each less opposite their where
    and either like or them whether
    another enough little our these which
    any every lots outside they while
    anybody everybody many over this who
    anyone everyone me own those whoever
    anything everything more past though whom
    are few most per through whose
    around following much plenty till will
    as for must plus to with
    at from my regarding toward within
    be have near same towards without
    because he need several under worth
    before her neither she unless would
    behind him no should unlike yes
    below i nobody since until you
    beside if none so up your
    """

    functional_words = functional_words.split()
    words = RemoveSpecialCHs(text)
    count = 0

    for i in text:
        if i in functional_words:
            count += 1

    return count / len(words)

dfinal['Functional_Words'] = dfinal.apply (lambda row: CountFunctionalWords(row.summary), axis=1)
dfinal.head()
```

```
def typeTokenRatio(text):
    words = word_tokenize(text)
    return len(set(words)) / len(words)
  
dfinal['type_token_ratio'] = dfinal.apply (lambda row: typeTokenRatio(row.summary), axis=1)
dfinal.head()
```
